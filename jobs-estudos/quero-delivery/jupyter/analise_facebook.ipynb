```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline

# load data
qd_df = pd.read_csv('../datas/Data-analysis-traffics-facebook281221.csv',
                    encoding='unicode_escape')

#Table manipulation:
qd_df.drop([
    'Nível de veiculação', 'Tipo de resultado', 'Término dos relatórios',
    'Início dos relatórios'
],
           axis=1,
           inplace=True)
qd_df['Dia'] = qd_df['Dia'].str.replace('-', '').astype(np.datetime64)
qd_df['Dia'] = pd.to_datetime(qd_df['Dia']).dt.strftime('%d/%m/%y')

qd_df.insert(4, 'Weekday', pd.to_datetime(qd_df['Dia']).dt.strftime('%A'))
qd_df['Weekday'] = qd_df['Weekday'].map({
    'Monday': 'Segunda-feira',
    'Tuesday': 'Terça-feira',
    'Wednesday': 'Quarta-feira',
    'Thursday': 'Quinta-feira',
    'Friday': 'Sexta-feira',
    'Saturday': 'Sábado',
    'Sunday': 'Domingo'
})

#Adsets rename
qd_df['Nome do conjunto de anúncios'] = qd_df[
    'Nome do conjunto de anúncios'].map(
        {'QD_Natal||All_cities||traffic': 'QD_Natal',
            'QD_Produtos||All_cities||traffic':'QD_Produtos' })
```

# Principais métricas por grupo de anúncio:

```python
adsetsum= qd_df.groupby('Nome do conjunto de anúncios').sum()
#series manipulation from adsetsum
adsetsum['Valor gasto (BRL)'] = adsetsum['Valor gasto (BRL)'].round(2).map(
    'R${}'.format)
#-----
adset1= adsetsum[['Valor gasto (BRL)','Alcance', 'Resultados',
    'Ativações de aplicativos','Ativações do aplicativo únicas']] 

adsetmean = qd_df.groupby('Nome do conjunto de anúncios').mean()
#series manipulation from adsetmean
adsetmean['Custo por resultado'] = adsetmean['Custo por resultado'].round(2).map(
    'R${}'.format)
adsetmean['Custo por ativação de aplicativo'] = adsetmean['Custo por ativação de aplicativo'].round(2).map(
    'R${}'.format)
adsetmean['Custo por ativação do aplicativo única'] = adsetmean['Custo por ativação do aplicativo única'].round(2).map(
    'R${}'.format)
#-----

adset2= adsetmean[['Custo por resultado','Custo por ativação de aplicativo',
   'Custo por ativação do aplicativo única']]

#concat------
adset_final= pd.concat([adset1,adset2],axis=1, join='inner')

adset_final[[
    'Valor gasto (BRL)', 'Alcance', 'Resultados', 'Custo por resultado',
    'Ativações de aplicativos', 'Custo por ativação de aplicativo',
    'Ativações do aplicativo únicas', 'Custo por ativação do aplicativo única'
]].sort_values(by = 'Ativações do aplicativo únicas', ascending = False)
```

```python
all_clicks = qd_df['Cliques (todos)'].sum()
min_cost_per_click = qd_df['Custo por resultado'].mean()
total_cost = qd_df['Valor gasto (BRL)'].sum()
all_activate = qd_df['Ativações de aplicativos'].sum()
cost_per_activate = qd_df['Custo por ativação do aplicativo única'].mean()
best_city_result = qd_df.groupby('Região').sum()


best_weekday = qd_df.groupby('Weekday').sum()

best_date = qd_df.groupby('Dia').sum()
mask_result1= (best_date['Ativações de aplicativos'] == best_date['Ativações de aplicativos'].max())
best_date_result = best_date[mask_result1]
```

## Total de cliques em todas as regiões

```python
plot_style = plt.style.use(['bmh'])
fig = plt.figure(figsize=(15,6))

set_graph_region= qd_df.groupby('Região')['Cliques (todos)'].sum().plot(kind = 'bar', width=0.8)
plt.title('Qtd De cliques por Região')

for i in set_graph_region.patches:
    plt.annotate(i.get_height().round(2),
            (i.get_x() + i.get_width()/ 2, i.get_height()),
            ha='center', va = 'baseline',
            fontsize = 10, xytext = (0,1), textcoords = 'offset points')
set_graph_region.set_xlabel( "Regiões", fontsize = 16)
set_graph_region.set_ylabel('Total de cliques', fontsize = 14)



```

```python
plt.style.use(['bmh'])
fig = plt.figure(figsize=(15,10))
gs = gridspec.GridSpec(3,2)
order_day = [ 'Domingo','Segunda-feira', 'Terça-feira', 'Quarta-feira', 'Quinta-feira', 'Sexta-feira', 'Sábado']
set_graph = qd_df
ax1 = fig.add_subplot(221)
ax1_graph = set_graph.groupby('Weekday') ['Resultados'].sum().loc[order_day].plot(kind='barh', 
        width = 0.8, color = 'springgreen',
        title = 'Resultados por dia da semana')
for i in ax1_graph.patches:
  plt.annotate(i.get_width(),
            ( i.get_width()+0.5, i.get_y()+i.get_height() /2),
            fontsize = 10
                 )
ax1_graph.set_ylabel('')
#_______

ax2 = fig.add_subplot(222)
ax2_graph= set_graph.groupby('Weekday') ['Custo por resultado'].mean().loc[order_day].plot(kind = 'bar', 
       width = 0.8, color = 'orangered', title = 'Custo médio por dia da semana')
for i in ax2_graph.patches:
  plt.annotate('R$'+str(i.get_height().round(2)),
            (i.get_x() + i.get_width()/ 2, i.get_height()),
            ha='center', va = 'baseline',
            fontsize = 10, xytext = (0,1), textcoords = 'offset points' )
ax2_graph.set_xlabel('')
#______

ax3 = fig.add_subplot(gs[-1,:])
ax3_graph = set_graph.groupby('Weekday') ['Frequência'].mean().loc[order_day].plot( marker='o', ms=10,  mec='b',color = 'red',
                    title = 'Frequência por dia da semana')
ax3_graph.set_xlabel('')
```

No gráfico acima, observe que o melhor dia não foi o que houve a maior frequência de alcance, ou seja, não necessáriamente os anúncios do Quero delivery precisam aparecer mais que 2 vezes para obter bons resultados de engajamento com a campanha.
